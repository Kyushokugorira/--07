## **プログラムの名称** RPGの戦闘
敵とエンカウントして戦闘処理をするプログラムになっています。
敵とのエンカウントや攻撃のパターンの確率を調整しています。
## **操作方法**
メイン処理のプログラムで攻撃をする場合,半角のaを入力,防御をする場合,半角のbを入力,魔法を使う場合,半角のcを入力が必要になります。
## **制作者のコメント**
確率や設定を変更しやすいように制作しました。
勇者が戦闘中最初に魔法ヒールを使うと元のHPより増えるようになっています。
戦闘中の処理をどのようにするか悩みました。
設定を考えるのが面白かったので、プログラムを楽しく作成することができました。

```python
#@title サブ処理
#@markdown **■ プレイヤーネーム**（日本語設定可）
player_name = '\u9D8F\u5C71\u5C0F\u8C46\u90CE' #@param {type:'string'}
player_age=53#@param {type:'integer'}

#@markdown **■ 難易度**
level = 'Easy' #@param ['Easy', 'Normal','Hard']
if level=='Hard':
  lvxp=10
elif level=='Normal':
  lvxp=5
else:
  lvxp=2
# 各種ライブラリのインポート
import random as r
# データやパラメータ設定
yuusya={ # 各ペアはカンマで区切る
  'level': 1,
  'yh': 50,
  'yad': 3,
  'ymd': 2,
  'mp':10,
  '魔法': 'ヒール',
  '称号': '一般人', # key には日本語も利用可能
  'items': ['いざないの剣','たびびとのふく','やくそう'], # リストも格納可能
  '経験値': 0
}
slime={
  'mn': 'スライム',
  'level': 1,
  'mh': 50,
  'mad': 2,
  'mmd': 0,
  'balance':[1,0],
  '称号': '巨大モンスター', # key には日本語も利用可能
  'items': ['水'], # リストも格納可能
  '報酬経験値': 2
}
goblin={
  'mn': 'ゴブリン',
  'level': 2,
  'mh': 60,
  'mad': 3,
  'mmd': 0,
  'balance':[1,0],
  '称号': '小型', # key には日本語も利用可能
  'items': ['皮','棍棒'], # リストも格納可
  '報酬経験値': 5
}
thief={
  'mn': '盗賊',
  'level': 5,
  'mh': 50,
  'mad': 10,
  'mmd': 5,
  'balance':[2,1],
  '魔法':'氷の雨',
  '称号': '盗賊', # key には日本語も利用可能
  'items': ['皮','ナイフ'], # リストも格納可能
  '報酬経験値': 5
}
soldier={
  'mn': '兵士',
  'level': 10,
  'mh': 100,
  'mad': 50,
  'mmd': 0,
  'balance':[1,0],
  '称号': '上級兵士', # key には日本語も利用可能
  'items': ['鎧','聖剣'], # リストも格納可能
  '報酬経験値': 50
}
fake={
  'mn': f'偽勇者 {player_name}',
  'level': 10,
  'mh': 200,
  'mad': 100,
  'mmd': 100,
  'balance':[10,3],
  '魔法':'フレイム',
  '称号': '上級兵士', # key には日本語も利用可能
  'items': ['鎧','聖剣'], # リストも格納可能
  '報酬経験値': 100
}
maou={
  'mn': f'魔王',
  'level': 200,
  'mh': 10000,
  'mad': 500,
  'mmd': 1000,
  'balance':[2,8],
  '称号': 'ラスボス', # key には日本語も利用可能
  '魔法':'闇の魔術',
  'items': ['魔王城'], # リストも格納可能
  '報酬経験値': 0
}
```

```python
#@title メイン処理
mn=[slime,goblin,thief,soldier,fake,maou]
mp=100,60,5,2,2,1
c=r.choices(mn,weights=map(int,mp),k=1)
sm=c[0]
mh=sm['mh']
md=sm['mad']
mmd=sm['mmd']
p=player_name
old=player_age
yh=yuusya['yh']
yad=yuusya['yad']
ym=yuusya['mp']
#関数の定義
def monsterattack():
  global yh
  global md
  global mmd
  dchoice=['attack','magic']
  if sm['mn']=='魔王':
    e=2
  else:
    e=1
  for i in range(e):
    d=r.choices(dchoice,weights=sm['balance'],k=1)
    if d[0]=='magic':
      print(f"{sm['mn']}の魔法 {sm['魔法']} が発動!")
      print(f"勇者{p}に{mmd}のダメージ! 勇者 {p}のHP {yh} -> {yh-mmd}")
      yh-=mmd
    elif d[0]=='attack':
      print(f"{sm['mn']}の攻撃、勇者 {p}に{md}のダメージ! 勇者 {p}のHP {yh} -> {yh-md}")
      yh-=md
#エンカウント
print(f"勇者 {p}の前に{sm['mn']}があらわれた。")
print(f"勇者 {p}のHP「{yh}」、{sm['mn']}のHP「{mh}」。")
#戦闘
while yh>0<mh:
  print(">>> 行動を選択してください")
  print(">>> a:攻撃 b:防御 c:魔法")
  action=input(">>> ")
  if action=="a":
    print(f"勇者 {p}の攻撃、{sm['mn']}に{yad}のダメージ! {sm['mn']}のHP {mh} -> {mh-yad}")
    mh-=yad
    if mh>0:
      monsterattack()
  elif action=="b":
    print("脳筋勇者に防御という言葉はない")
    print(f"勇者 {p}は防御の仕方がわからなかった")
    monsterattack()
  elif action=="c":
    if ym>2:
      print(f"勇者 {p}は魔法{yuusya['魔法']}を発動!")
      print(f"MPを2消費してHPを10回復させた! 勇者 {p}のHP {yh} -> {yh+10} 勇者 {p}のMP {ym} -> {ym-2}")
      ym-=2
      yh+=10
      monsterattack()
    else:
      print("MPが足りません")
  else :
    print("半角a,b,cのどれか一文字を入力してください")
#戦闘終了後処理
if yh<=0:
  print()
  print(f"享年{old}年 勇者 {p} 死亡")
else:
  yuusya['経験値']+=sm['報酬経験値']
  print("戦闘終了!")
  print(f"経験値 {sm['報酬経験値']}")
  print(f"{','.join(sm['items'])}を手に入れた")
  yuusya['items'].extend(sm['items'])
  while yuusya['経験値']>=lvxp:
    print(f"勇者{p}のレベルが上がった! 勇者 {p}のlevel {yuusya['level']+1}")
    yuusya['経験値']-=lvxp
    yuusya['level']+=1
    yuusya['yh']+=10
    yuusya['yad']+=2
  if sm['mn']=='魔王':
    print('ゲームクリア')
```
